```python
import numpy as np

# Load the .npy file
elevation_data = np.load('output/CA3982012144020181108/elevation.npy', allow_pickle=True)

# Show the content
print(elevation_data)
```

```python
import matplotlib.pyplot as plt

# Access the dictionary inside the 0-d array
elevation_dict = elevation_data.item()

# Plot the elevation data
plt.figure(figsize=(10, 8))
plt.imshow(elevation_dict['data'], cmap='terrain')
plt.colorbar(label='Elevation (m)')
plt.show()
```

```python
# Load the fire data .npy file
fire_data = np.load('output/CA3982012144020181108/fire_data.npy', allow_pickle=True)

# Access the dictionary inside the 0-d array
fire_dict = fire_data.item()

# Plot the fire data
plt.figure(figsize=(10, 8))
plt.imshow(fire_dict['data'][0], cmap='hot', interpolation='nearest')
plt.colorbar(label='Fire Intensity')
plt.show()
```
